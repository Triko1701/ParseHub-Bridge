```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException
from models import job_details_selector, job_listing_selector # JobListingSelector, JobDetailsSelector, JobDetails
from postgresql_client import PostgresqlClient
    
def get_post_details(driver, job_url_list, JOB_LISTING_TAB, JOB_DETAILS_TAB, max_wait_time: int=20,
                     db = "job", tb = "job_listings", password = "alex"):
    driver.switch_to.window(JOB_DETAILS_TAB)
    
    pg_client = PostgresqlClient(db = db, tb = tb, password = password)
    
    for job_url in job_url_list:
        driver.get(job_url)
        try:
            WebDriverWait(driver, max_wait_time).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, job_details_selector.title))
            )
            print("Found job title")
        except TimeoutException:
            print(f"Job title not found after {max_wait_time}s. Page may not have loaded correctly!")
            
        # Iterate through the attributes
        
        job_details = dict()
        job_details["url"] = job_url
        for key, value in job_details_selector.model_dump().items():
            try:
                job_details[key] = driver.find_element(By.CSS_SELECTOR, value).text
            except:
                job_details[key] = ""
            
            print(f"{key}: {job_details[key]}")
            
        # SQL command to insert a new row using the dictionary keys as columns
        insert_query = """
        INSERT INTO {} ({}) VALUES ({})
        """.format(
            tb,
            ', '.join(job_details.keys()),
            ', '.join(['%({})s'.format(k) for k in job_details.keys()])
        )
        pg_client.query(query=insert_query, params=job_details, ret=False)


    driver.switch_to.window(JOB_LISTING_TAB)


def JobScraping(url: str,
                job_listing_selector = job_listing_selector,
                job_details_selector = job_details_selector,
                PATH_TO_CHROMEDRIVER: str='chromedriver-win64/chromedriver.exe',
                max_wait_time: int=20
                ):


    # Create a Service object with the path to the chromedriver
    service = Service(PATH_TO_CHROMEDRIVER)

    # Initialize the Chrome WebDriver with the Service object
    driver = webdriver.Chrome(service=service)
    driver.execute_script("window.open('');")

    JOB_LISTING_TAB = driver.window_handles[0]
    JOB_DETAILS_TAB = driver.window_handles[1]

    driver.switch_to.window(JOB_LISTING_TAB)
    driver.get(url)

    # while not no_result && pagination_element && next_button:
    while True:
        # Check "no result" error message
        try:
            # no_result_element = driver.find_element(By.CSS_SELECTOR, job_listing_selector.no_result)
            no_result_element = WebDriverWait(driver, max_wait_time).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, job_listing_selector.no_result))
            )
            no_result = "We couldn't find anything that matched your search." in no_result_element.text
            if no_result:
                print("No results found!")
                break
        except:
            print("Results found!")
            
        # Wait 10s for the page to load
        try:
            WebDriverWait(driver, max_wait_time).until(
                EC.visibility_of_element_located((By.CSS_SELECTOR, job_listing_selector.pagination))
            )
            print("pagination_selector element found!")
            
        except TimeoutException:
            print(f"Pagination element not found after {max_wait_time}s. Page may not have loaded correctly!")
            # break
            
        # Find all <a> tags with the specific 'data-automation' attribute
        try:
            job_links_elements = driver.find_elements(By.CSS_SELECTOR, job_listing_selector.job_link)
        except:
            print("No job links found!")
            break

        # Extract URLs from the elements
        job_url_list = [element.get_attribute('href') for element in job_links_elements if element.get_attribute('href')]
        # print(job_url_list)
        print("No. of job links found: {}".format(len(job_url_list)))
        # Get post details
        get_post_details(driver, job_url_list, JOB_LISTING_TAB, JOB_DETAILS_TAB)
            
        # Check if the 'Next' button is present and clickable
        try:            
            next_button = WebDriverWait(driver, max_wait_time).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, job_listing_selector.next_button))
            )
            
            if next_button:
                next_button.click()
                
        except TimeoutException:
            print(f"No next button found after {max_wait_time}s. Last page might have been reached!")
            break 

url = "https://www.seek.com.au/jobs-in-information-communication-technology/engineering-hardware"
JobScraping(url = url)

  
```
